## ODES

**Euler N = 1000**

```python
import numpy as np
import matplotlib.pyplot as plt
import odes as od

r0 = np.array([1,0])
N = 1000
e = od.euler(r0,N)
t = np.arange(0,5*(2*np.pi),2*np.pi/N)
eplot= od.eplot(e,t)
```

**Smallest acurate time step N = 600 Euler**

```python
N = 600
e = od.euler(r0,N)
t = np.arange(0,5*(2*np.pi),2*np.pi/N)
eplot= od.eplot(e,t)
```

**inaccurate time step N = 100 Euler**

**Heun N = 1000**

```python
r0=np.array([1,0])
N=1000
h = od.heun(r0,N)
t = np.arange(0,5*(2*np.pi),2*np.pi/N)
hplot= od.hplot(h,t)
```

## Observations

The Euler metod seems like it takes the greatest number of time steps to get an accurate graph. The Heun takes the sencond greatest number of time step, the 2nd Runge Kutta the third most and the 4th order Runge Kutta the least number of time steps to get an accurate depiction of the graphs. The smallest accurate N for Euler was about 600, the smallest N for Heun was about 50, the smallest accurate N for 2nd Runge Kutta was about 30, and for 4th Runge Kutta it was about 9.
